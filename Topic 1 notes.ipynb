```python
import numpy as np
import pandas as pd

pd.set_option("display.max.columns", 100)
# to draw pictures in jupyter notebook
%matplotlib inline
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings

import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
```

```python
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/"
```

```python
data = pd.read_csv(DATA_URL + "adult.data.csv")
data.head()
```

```python
# 1. How many men and women (sex feature) are represented in this dataset?

data["sex"].value_counts()
```

```python
# 2. What is the average age (age feature) of women?

data[data["sex"] == "Female"]["age"].mean()
```

```python
# 3. What is the percentage of German citizens (native-country feature)?

german_count = (data["native-country"] == "Germany").sum()
percent = (german_count / data.shape[0]) * 100
percent
```

```python
# 4-5. What are the mean and standard deviation of age for those who earn more 
# than 50K per year (salary feature) and those who earn less than 50K per year?

ages_high = data[data["salary"] == ">50K"]["age"]
agas_low = data[data["salary"] == "<=50K"]["age"]

print(f'>50K mean age = {(ages_high.mean()).round(2)}')
print(f'>50K standard deviation age = {(ages_high.std()).round(2)}')
print(f'<=50K mean age = {(agas_low.mean()).round(2)}')
print(f'<=50K standard deviation age = {(agas_low.std()).round(2)}')
```

```python
# 6. Is it true that people who earn more than 50K have at least high school 
# education? (education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, 
# Masters or Doctorate feature) - No

data[data["salary"] == ">50K"]["education"].unique()
```

```python
# 7. Display age statistics for each race (race feature) and each gender (sex 
# feature). Use groupby() and describe(). Find the maximum age of men of 
# Amer-Indian-Eskimo race. - 82

# data.groupby(["race"])["age"].describe(percentiles=[])

for (race, sex), sub_df in data.groupby(["race", "sex"]):
    print("Race: {0}, sex: {1}".format(race, sex))
    print(sub_df["age"].describe())
```

```python
# 8. Among whom is the proportion of those who earn a lot (>50K) greater: 
# married or single men (marital-status feature)? Consider as married those 
# who have a marital-status starting with Married (Married-civ-spouse, 
# Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

married_man = data[(data["sex"] == "Male")
    & (data["marital-status"].str.startswith("Married"))][
    "salary"].value_counts(normalize=True)

not_married_man = data[(data["sex"] == "Male")
    & ~(data["marital-status"].str.startswith("Married"))
    ]["salary"].value_counts(normalize=True)

print(f'married man: \n{married_man}')
print(f'not married man: \n{not_married_man}')
```

```python
# 9. What is the maximum number of hours a person works per week 
# (hours-per-week feature)? How many people work such a number of hours, and 
# what is the percentage of those who earn a lot (>50K) among them?

max_hours = data["hours-per-week"].max()
print(f"max hours per week {max_hours}")

max_hours_sum = (data["hours-per-week"] == max_hours).sum()
print(f"max hours per week {max_hours_sum}")

rich_share = (
    float(
        data[(data["hours-per-week"] == max_hours) 
        & (data["salary"] == ">50K")].shape[0]) / max_hours_sum
) * 100

print(f"Percents of rich among them = {rich_share.round(2)}")
```

```python
# 10. Count the average time of work (hours-per-week) for those who earn a 
# little and a lot (salary) for each country (native-country). What will these 
# be for Japan?

for (country, salary), sub_df in data.groupby(["native-country", "salary"]):
    print(country, salary, round(sub_df["hours-per-week"].mean(), 2))
```
